## Construct Neural Network

```python
import torch.nn.functional as F
```

```python
class Net(torch.nn.Module):    # inherit torch Module
    def __init__(self,n_feature,n_hidden,n_output):  
        super(Net,self).__init__()         # inherit     __init__
        # define the output of each layer
        self.hidden = torch.nn.Linear(n_feature,n_hidden)     # linear output of hidden layer
        self.predict = torch.nn.Linear(n_hidden,n_output)     # linear output of output layer
    def forward(self,x):
        x = F.relu(self.hidden(x))     # activation function
        x=self.predict(x)           # output
        return x
```

```python
net=Net(n_feature=1,n_hidden=10,n_output=1)
print(net)
```

## Train Neural Network

```python
x=torch.unsqueeze(torch.linspace(-1,1,100),dim=1)
y=x.pow(2)+0.2*torch.rand(x.size())

optimizer= torch.optim.SGD(net.parameters(),lr=0.5)
loss_func=torch.nn.MSELoss()  # calculate the error between estimated and real value

for t in range(100):
    prediction=net(x)
    loss=loss_func(prediction,y)
    
    optimizer.zero_grad()     # clean the parameters in last step
    loss.backward()         # calculate the updated loss
    optimizer.step()        # new parameters to net
```

## save

```python
import torch
from torch.autograd import Variable 
torch.manual_seed(1)

x=torch.unsqueeze(torch.linspace(-1,1,100),dim=1)
y=x.pow(2)+0.2*torch.rand(x.size())  

```

```python
def save():
    # construct
    net1=torch.nn.Sequential(torch.nn.Linear(1,10),
                            torch.nn.ReLU(),
                            torch.nn.Linear(10,1))
    optimizer= torch.optim.SGD(net1.parameters(),lr=0.5)
    loss_func=torch.nn.MSELoss() 
    # train
    for t in range(100):
        prediction = net1(x)
        loss = loss_func(prediction, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step() 
    torch.save(net1, 'net.pkl')  # save the whole NN
    torch.save(net1.state_dict(), 'net_params.pkl')     # save the parameters，faster, less GPU
    
    # plot result
    plt.figure(1,figsize=(10,3))
    plt.subplot(131)
    plt.title('net1')
    plt.scatter(x.data.numpy(),y.data.numpy())
    plt.plot(x.data.numpy(),prediction.data.numpy(),'r-',lw=5)
    
    
# extract whole NN
def restore_net():
    #restore entire net1 to net2
    net2=torch.load("net.pkl")
    prediction=net2(x)
    
    # plot result
    plt.figure(1,figsize=(10,3))
    plt.subplot(132)
    plt.title('net2')
    plt.scatter(x.data.numpy(),y.data.numpy())
    plt.plot(x.data.numpy(),prediction.data.numpy(),'r-',lw=5)
    
    
    
# extract parameters
def restore_params():
    net3=torch.nn.Sequential(
    torch.nn.Linear(1,10),
    torch.nn.ReLU(),
    torch.nn.Linear(10,1)
    )
    
    # duplicate the parameters to net3
    net3.load_state_dict(torch.load('net_params.pkl'))
    prediction=net3(x)
    
 
    # plot result
    plt.figure(1,figsize=(10,3))
    plt.subplot(133)
    plt.title('net3')
    plt.scatter(x.data.numpy(),y.data.numpy())
    plt.plot(x.data.numpy(),prediction.data.numpy(),'r-',lw=5)
    plt.show()
```

```python
save()
```

```python
restore_net()
```

```python
restore_params()
```

# Classification

```python
import torch
import matplotlib.pyplot as plt
```

```python
n_data=torch.ones(100,2)
x0=torch.normal(2*n_data,1)    # type0 x data(tensor), shape=(100,2)
y0=torch.zeros(100)            # type0 y data(tensor), shape=(100,)
x1=torch.normal(-2*n_data,1)   # type1 x data(tensor), shape=(100,2)
y1=torch.ones(100)             # type1 y data(tensor), shape=(100,)
```

torch.cat(tensors, dim=0, *, out=None) → Tensor

```python
x=torch.cat((x0,x1),0).type(torch.FloatTensor)
y=torch.cat((y0,y1),).type(torch.LongTensor)
```

```python
x.shape
```

```python
y.shape
```

```python
plt.scatter(x.data.numpy()[:,0],x.data.numpy()[:,1],c=y.data.numpy(),s=100,lw=0,cmap='RdYlGn')
plt.show()
```

```python
import torch
import torch.nn.functional as F

class Net(torch.nn.Module):    # inherit torch Module
    def __init__(self,n_feature,n_hidden,n_output):  
        super(Net,self).__init__()         # inherit     __init__
        # define the output of each layer
        self.hidden = torch.nn.Linear(n_feature,n_hidden)     # linear output of hidden layer
        self.out = torch.nn.Linear(n_hidden,n_output)     # linear output of output layer
    def forward(self,x):
        x = F.relu(self.hidden(x))     # activation function
        x=self.out(x)           # output
        return x
```

```python
net=Net(n_feature=2,n_hidden=10,n_output=2)
net
```

```python
# train
optimizer=torch.optim.SGD(net.parameters(),lr=0.02)
loss_func=torch.nn.CrossEntropyLoss()
for t in range(100):
    out=net(x)
    loss=loss_func(out,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
import matplotlib.pyplot as plt
plt.ion()
plt.show()
for t in range(100):
    out=net(x)
    loss=loss_func(out,y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if t % 2 == 0:
        plt.cla()
        prediction = torch.max(F.softmax(out), 1)[1]
        pred_y = prediction.data.numpy().squeeze()
        target_y = y.data.numpy()
        plt.scatter(x.data.numpy()[:, 0], x.data.numpy()[:, 1], c=pred_y, s=100, lw=0, cmap='RdYlGn')
        accuracy = sum(pred_y == target_y)/200. 
        plt.text(1.5, -4, 'Accuracy=%.2f' % accuracy, fontdict={'size': 20, 'color':  'red'})
        plt.pause(0.1)
plt.ioff()
plt.show()
```
